```python
class CartItem():
    def __init__(self, name="noname", price=0, quantity=0):
        self.__productName = name
        self.__quantity = quantity
        self.__price = price

    @property
    def productName(self):
        return self.__productName

    @productName.setter
    def productName(self, name):
        if isinstance(name, str) and len(name.strip()) > 0:
            self.__productName = name.strip()
        else:
            print("Недопустимое имя товара. Имя должно быть непустой строкой.")

    @property
    def quantity(self):
        return self.__quantity

    @quantity.setter
    def quantity(self, quantity):
        if isinstance(quantity, int) and quantity >= 0:
            self.__quantity = quantity
        else:
            print("Недопустимое количество. Количество должно быть неотрицательным целым числом.")

    @property
    def price(self):
        return self.__price

    @price.setter
    def price(self, price):
        if isinstance(price, (int, float)) and price >= 0:
            self.__price = price
        else:
            print("Недопустимая цена. Цена должна быть неотрицательным числом.")
item = CartItem("Laptop", 50000, 2)
print(f"Имя товара: {item.productName}, Цена: {item.price}, Количество: {item.quantity}")

item.productName = "Ноутбук"
item.price = 60000
item.quantity = 3
print(f"Имя товара: {item.productName}, Цена: {item.price}, Количество: {item.quantity}")

item.price = -100
item.quantity = -5
item.productName = ""
```

```python
class Shelf:
    def __init__(self, cabinet):
        self.cabinet = cabinet
        self.items = []

    def add_item(self, item):
        if isinstance(item, str) and item.strip():
            self.items.append(item.strip())
        else:
            print("Недопустимый предмет. Предмет должен быть непустой строкой.")

    def __str__(self):
        if not self.items:
            return "Полка пустая"
        result = "Полка с предметами: "
        for i, item in enumerate(self.items):
            if i > 0:
                result += ", "
            result += item
        return result


class Cabinet:
    def __init__(self):
        self.shelves = []

    def add_shelf(self):
        new_shelf = Shelf(self)
        self.shelves.append(new_shelf)
        return new_shelf

    def count_shelves(self):
        return len(self.shelves)

    def __str__(self):
        return (f"Шкаф с {self.count_shelves()} полками")


cabinet = Cabinet()
shelf1 = cabinet.add_shelf()
shelf2 = cabinet.add_shelf()

shelf1.add_item("Книга")
shelf1.add_item("Чашка")
shelf2.add_item("Ваза")

print(cabinet)
print(shelf1)
print(shelf2)
```

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def __str__(self):
        return (f'Название "{self.title}", Автор: {self.author}')


class Reader:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"{self.name}"


class Library:
    def __init__(self):
        self.books = []
        self.readers = []

    def add_book(self, book):
        if isinstance(book, Book):
            self.books.append(book)
        else:
            print("Недопустимая книга. Ожидается объект класса Book.")

    def add_reader(self, reader):
        if isinstance(reader, Reader):
            self.readers.append(reader)
        else:
            print("Недопустимый читатель. Ожидается объект класса Reader.")

    def __str__(self):
        books_info = "Книги:\n"
        for book in self.books:
            books_info += str(book) + "\n"

        readers_info = "Читатели:\n"
        for reader in self.readers:
            readers_info += str(reader) + "\n"

        return f"Библиотека:\n{books_info}{readers_info}"


book1 = Book("1984", "Джордж Оруэлл")
book2 = Book("Мастер и Маргарита", "Михаил Булгаков")

reader1 = Reader("Иван Иванов")
reader2 = Reader("Александр Пушкин")

library = Library()
library.add_book(book1)
library.add_book(book2)
library.add_reader(reader1)
library.add_reader(reader2)

print(library)

```

```python
from abc import ABC, abstractmethod

class Shape2D(ABC):
    @abstractmethod
    def calculatePerimeter(self):
        pass

class Square(Shape2D):
    def __init__(self, side_length: float):
        self.side_length = side_length

    def calculatePerimeter(self):
        return 4 * self.side_length

    def __str__(self):
        return f"Квадрат: Длина стороны = {self.side_length}, Периметр = {self.calculatePerimeter()}"

class Triangle(Shape2D):
    def __init__(self, side1: float, side2: float, side3: float):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3

    def calculatePerimeter(self):
        return self.side1 + self.side2 + self.side3

    def __str__(self):
        return f"Треугольник: Стороны = {self.side1}, {self.side2}, {self.side3}, Периметр = {self.calculatePerimeter()}"

square = Square(5)
print(square)

triangle = Triangle(3, 4, 5)
print(triangle)
```

```python

class IProduct(ABC):
    @abstractmethod
    def getTotalValue(self):
        pass

    @abstractmethod
    def updatePrice(self, price: float):
        pass

class Product(IProduct):
    def __init__(self, name: str, price: float, quantity: int):
        self.name = name
        self.price = price
        self.quantity = quantity

    def getTotalValue(self):
        return self.price * self.quantity

    def updatePrice(self, price: float):
        if price >= 0:
            self.price = price
        else:
            print("Цена не может быть отрицательной.")

    def __str__(self):
        return f"Продукт: {self.name}, Цена: {self.price}, Количество: {self.quantity}"



product = Product("Ноутбук", 50000, 3)
print(product)
print(f"Общая стоимость: {product.getTotalValue()}")

product.updatePrice(45000)
print(product)
print(f"Общая стоимость после обновления цены: {product.getTotalValue()}")

product.updatePrice(-100)
```
