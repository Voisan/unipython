```python
def word_frequency(text):
    words = text.split()
    return {word: words.count(word) for word in words}
```

```python
def max_key(d):
    return max(d, key=d.get)
```

```python
def unique_values(d):
    return set(d.values())
```

```python
def sum_values(d):
    return sum(d.values())
```

```python
def remove_keys(d, keys):
    return {k: v for k, v in d.items() if k not in keys}
```

```python
def keys_with_even_values(d):
    return [k for k, v in d.items() if v % 2 == 0]
```

image.png

```python
def sort_by_keys(d):
    return dict(sorted(d.items()))
```

```python
def group_by_value_types(dictionary):
    result = {}
    for key, value in dictionary.items():
        value_type = type(value).__name__
        if value_type not in result:
            result[value_type] = []
        result[value_type].append(key)
    return result
```

```python
def value_frequency(dictionary):
    frequency = {}
    for value in dictionary.values():
        if value in frequency:
            frequency[value] += 1
        else:
            frequency[value] = 1
    return frequency
```

```python
test_dict = {'a': 1, 'b': 2, 'c': 2, 'd': 3, 'e': 1}
test_text = "apple banana apple cherry banana apple"
test_keys_to_remove = ['a', 'c']

print("1. word_frequency:")
freq = word_frequency(test_text)
print(freq)

print("\n2. max_key:")
print(max_key(test_dict))

print("\n3. min_key:")
print(min_key(test_dict))

print("\n4. unique_values:")
print(unique_values(test_dict))

print("\n5. sum_values:")
print(sum_values(test_dict))

print("\n6. remove_keys:")
print(remove_keys(test_dict, test_keys_to_remove))

print("\n7. keys_with_even_values:")
print(keys_with_even_values(test_dict))

print("\n8. sort_by_keys:")
print(sort_by_keys({'c': 1, 'a': 2, 'b': 3}))

print("\n9. group_by_value_types:")
mixed_dict = {'a': 1, 'b': 'text', 'c': [1,2], 'd': 3.14}
print(group_by_value_types(mixed_dict))

print("\n10. value_frequency:")
print(value_frequency(test_dict))
```

```python
def list_to_tuple(lst):
    return tuple(lst)
```

image.png

```python
def is_in_tuple(t, element):
    return element in t
```

image.png

```python
def count_element(t, element):
    return t.count(element)
```

image.png

```python
def tuple_min(t):
    return min(t) if t else None
```

image.png

```python
def split_tuple(t):
    mid = len(t) // 2
    return (t[:mid], t[mid:])
```

image.png

```python
def tuple_slice(t, start, end):
    return t[start:end]
```

image.png

```python
def unique_in_tuple(t):
    return set(t)
```

image.png

```python
def filter_tuple(t, condition):
    #condition: функция-условие
    return tuple(x for x in t if condition(x))
```

image.png

```python
def tuple_product(t):
    product = 1
    for num in t:
        if type(num) in (int, float):
            product *= num
    return product
```

image.png

```python
def make_tuple_from_range(start, end):
    return tuple(range(start, end))
```

```python
test_list = [1, 2, 3]
test_tuple = (1, 2, 2, 3, 4)
mixed_tuple = (1, 'a', 2.5, 3, 'b')
range_tuple = (5, 1, 3, 8, 2)

print("1. list_to_tuple:")
print(list_to_tuple(test_list))

print("\n2. is_in_tuple:")
print(is_in_tuple(test_tuple, 2))
print(is_in_tuple(test_tuple, 5))

print("\n3. count_element:")
print(count_element(test_tuple, 2))
print(count_element(test_tuple, 5))

print("\n4. tuple_min:")
print(tuple_min(test_tuple))
print(tuple_min(()))

print("\n5. split_tuple:")
print(split_tuple(test_tuple))
print(split_tuple((1, 2, 3, 4)))

print("\n6. tuple_slice:")
print(tuple_slice(test_tuple, 1, 4))

print("\n7. unique_in_tuple:")
print(unique_in_tuple(test_tuple))

print("\n8. filter_tuple:")
print(filter_tuple(mixed_tuple, lambda x: isinstance(x, int)))
print(filter_tuple(range_tuple, lambda x: x > 3))

print("\n9. tuple_product:")
print(tuple_product((2, 3, 4)))
print(tuple_product(mixed_tuple))

print("\n10. make_tuple_from_range:")
print(make_tuple_from_range(3, 8))
print(make_tuple_from_range(0, 3))
```
