```python
def count_vowels(text):
    vowels = "аеёиоуыэюя"
    count = 0
    for char in text.lower():
        if char in vowels:
            count += 1
    return count

# Пример использования (можно заменить на реальное стихотворение Пушкина)
poem = """У лукоморья дуб зелёный;
Златая цепь на дубе том:
И днём и ночью кот учёный
Всё ходит по цепи кругом;"""

print(f"гласных букв: {count_vowels(poem)}")
```

```python
def count_letters(text):
    letters = {'r': 0, 'k': 0, 't': 0}
    for char in text.lower():
        if char in letters:
            letters[char] += 1
    return letters

# Пример использования
string = "This is a test string with some letters like k, r, and t."
result = count_letters(string)

print(f"Количество вхождений:")
print(f"r: {result['r']}")
print(f"k: {result['k']}")
print(f"t: {result['t']}")
```

```python
def check_brackets(text):
    open_brackets = 0
    for char in text:
        if char == '(':
            open_brackets += 1
        elif char == ')':
            open_brackets -= 1
            if open_brackets < 0:
                return False
    return open_brackets == 0

# Примеры использования
test_string1 = "(a + b) * (c - d)"  # Сбалансировано
test_string2 = "((x + y) * z"       # Не сбалансировано

print(f"Строка 1: {'Сбалансировано' if check_brackets(test_string1) else 'Не сбалансировано'}")
print(f"Строка 2: {'Сбалансировано' if check_brackets(test_string2) else 'Не сбалансировано'}")
```

```python
import random

def file_operations(m, n, size=20):
    with open('f.txt', 'w') as f:
        for _ in range(size):
            num = random.randint(1, 100)
            f.write(f"{num}\n")

    with open('f.txt', 'r') as f, open('g.txt', 'w') as g:
        for line in f:
            num = int(line.strip())
            if num % m == 0 and num % n != 0:
                g.write(f"{num}\n")

m_value = 3
n_value = 5
file_operations(m_value, n_value)

print("Файл f.txt создан со случайными числами")
print(f"Числа, делящиеся на {m_value} и не делящиеся на {n_value}, записаны в g.txt")
```

```python
import random

def create_baggage_file(filename, num_passengers):
    with open(filename, 'w') as file:
        for _ in range(num_passengers):
            items = random.randint(1, 5)
            weight = round(random.uniform(5, 50), 2)
            file.write(f"{items} {weight}\n")

def check_special_passenger(filename):
    with open(filename, 'r') as file:
        for line in file:
            items, weight = line.strip().split()
            items = int(items)
            weight = float(weight)
            if items == 1 and weight < 30:
                return True
    return False

create_baggage_file("Bagazh.txt", 10)

has_special = check_special_passenger("Bagazh.txt")
print(f"Найден пассажир с одной вещью весом менее 30 кг: {'Да' if has_special else 'Нет'}")
```

```python
def count_word_occurrences(filename, target_word):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            text = file.read()
            words = text.lower().split()
            target_word = target_word.lower()
            count = words.count(target_word)
            return count
    except FileNotFoundError:
        return -1

text_filename = "test.txt"
search_word = "тест"

sample_text = "Текст создан для теста. Надеюсь код пройдет тест и выведет 2. ТЕСТ "
with open(text_filename, 'w', encoding='utf-8') as f:
    f.write(sample_text)

occurrences = count_word_occurrences(text_filename, search_word)

if occurrences == -1:
    print("Файл не найден")
elif occurrences == 0:
    print(f"Слово '{search_word}' не найдено в тексте")
else:
    print(f"Слово '{search_word}' встречается {occurrences} раз(а)")
```
